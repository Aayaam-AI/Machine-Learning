```python
category = list(train["airline_sentiment"])
```

```python
train_text = list(train["text"])
test_text = list(test["text"])
```

```python
from nltk.tokenize import word_tokenize
```

```python
train_words = [word_tokenize(i) for i in train_text]
test_words = [word_tokenize(i) for i in test_text]
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
from nltk.corpus import stopwords
import string
stop = set(stopwords.words('english'))
punc = string.punctuation
stop.update(punc)
from nltk import pos_tag
```

```python
from nltk.corpus import wordnet
def get_simple_pos(pos):
    if pos.startswith('J'):
        return wordnet.ADJ
    elif pos.startswith('V'):
        return wordnet.VERB
    elif pos.startswith('N'):
        return wordnet.NOUN
    elif pos.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def clean(words):
    output_words = []
    for w in words:
        if w.lower() not in stop and w.isalpha():
            pos = pos_tag([w]) # Pass the word without lowering in pos and pass it as array or it will give pos tag of each aplphabet of word
            clean_word = lemmatizer.lemmatize(w , pos = get_simple_pos(pos[0][1])) # 0 will give first tuple and 1 will give its second element which is its pos tag
            output_words.append(clean_word.lower())
    return output_words 
```

```python
clean_train = [clean(words) for words in train_words]
clean_test = [clean(words) for words in test_words]
```

```python
test_text = [" ".join(word) for word in clean_test]
train_text = [" ".join(word) for word in clean_train]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 2000)
train_freq = cv.fit_transform(train_text).todense()
test_freq = cv.transform(test_text).todense()
```

```python
cv.get_feature_names()
```

```python
def name_to_num(s):
    num = 0
    if s == 'positive':
        num = 1
    elif s == 'negative':
        num = -1
    else:
        num = 0
    return num
```

```python
y_train = list(train['airline_sentiment'].apply(name_to_num))
```

```python
from sklearn.naive_bayes import ComplementNB
from sklearn.naive_bayes import MultinomialNB 
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

## Using Multinomial Naive Bayes

```python
nb1 = MultinomialNB()
nb1.fit(train_freq,y_train)
y_pred1 = nb1.predict(test_freq)
```

```python
print("Score on Training Data : ",nb1.score(train_freq,y_train))
```

## Using Complement Naive Bayes

```python
nb2 = ComplementNB()
nb2.fit(train_freq,y_train)
y_pred2 = nb2.predict(test_freq)
```

```python
print("Score on Training Data : ",nb2.score(train_freq,y_train))
```

## Now convert predictions into text

```python
# To convert predicted values into string
def num_to_name(x):
    if x == 0:
        out = 'neutral'
    elif x == 1:
        out = 'positive'
    else:
        out = 'negative'
    return out
```

```python
pred_name_1 = [num_to_name(i) for i in y_pred1]
pred_name_2 = [num_to_name(i) for i in y_pred2]
```
